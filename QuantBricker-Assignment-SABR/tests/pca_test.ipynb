### pca on swap data

```python
swap_data_diff=swap_data.diff().dropna()
swap_data_diff.head()
```

```python
X = swap_data_diff.values.T  
X_bar = X - X.mean(axis=1, keepdims=True) 
```

```python
n = X_bar.shape[1]  # number of observations
p = 3
```

```python
C = X_bar @ X_bar.T
eigenvalues, eigenvectors = np.linalg.eigh(C)

# Sort descending
idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

U = eigenvectors[:, :p]

print("C shape:", C.shape)
print("U shape:", U.shape)
print("Explained variance:", (eigenvalues[:p] / eigenvalues.sum()).round(4))
```

```python
Y = U.T @ X_bar 
print("Y shape:", Y.shape) 
```

```python
X_reconstructed = U @ Y  # (N, n)
print("Reconstructed X shape:", X_reconstructed.shape)
```

```python
reconstruction_error = np.mean((X_bar - X_reconstructed) ** 2)
print(f"Reconstruction MSE: {reconstruction_error:.4f}")
```

```python
## use only high liquidity instruments to hedge
from numpy.linalg import lstsq

hedge_names = ['1Y1Y', '1Y5Y', '5Y5Y', '10Y10Y']

risk_hedge = risk_all.loc[:, :].copy()
risk_all.index = list(swap_portfolio.keys())
H = risk_all.loc[hedge_names].values.astype(float)  # (4, 15)

H_pc = H @ U  # (4, 3)

p_pc = pc_risk_matrix.values.flatten().astype(float)  # (3,)

w, residuals, rank, sv = lstsq(H_pc.T, -p_pc, rcond=None)

print("Hedge weights:")
for name, weight in zip(hedge_names, w):
    print(f"  {name}: {weight:.4f}")

print("\nResidual PC exposure:", (p_pc + H_pc.T @ w).round(6))
```
