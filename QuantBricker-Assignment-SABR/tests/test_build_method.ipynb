## Test Model Build Method

```python
### create yield curve index build method
build_method_type = 'YIELD_CURVE_INDEX'
content = {
    'TARGET' : 'SOFR-1B',
    'OVERNIGHT INDEX FUTURE' : 'SOFR-FUTURE-3M',
    'OVERNIGHT INDEX SWAP' : 'USD-SOFR-OIS',
    'INSTANTANEOUS FORWARD RATE' : 'USD-SOFR-IFR'
}
yc_build_method = qfCreateBuildMethod(build_method_type, content)
display(yc_build_method.display())
```

```python
### serialize / de-serialize
path = 'yc_build_method.pickle'
qfWriteBuildMethodToFile(yc_build_method, path)
display(yc_build_method.display())
yc_build_method_back = qfReadBuildMethodFromFile(path)
# check
display(yc_build_method_back.display())
# house keeping
os.remove(path)
```
