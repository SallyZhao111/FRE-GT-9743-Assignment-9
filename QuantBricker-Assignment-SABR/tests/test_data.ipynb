## Test Data 1D

```python
df = pd.DataFrame(index=['3M', '6M', '9M', '1Y'])
df['values'] = [.03] * len(df)
data_type = 'Overnight Index Future'
data_conv = 'SOFR-FUTURE-3M'
data1D = qfCreateData1D(data_type, data_conv, df)
display(data1D.display())
# serialize
path = 'data1d.pickle'
qfWriteDataObjectToFile(data1D, path)
# deserialize
data1D_back = qfReadDataObjectFromFile(path)
# check
display(data1D_back.display())
# house keeping
os.remove(path)
```
